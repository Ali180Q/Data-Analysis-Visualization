```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#import dataset titanic from seaborn
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter',x='age',y='fare',color='red')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
```

```python
#bar plot USING SEABORN
sns.barplot(data=df, x='survived',y='fare',hue='class')
```

```python
sns.boxplot(data=df, x='survived',y='fare',hue='class')
```
